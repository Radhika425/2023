```python
df.isnull().mean().plot.bar(figsize=(10,6))
plt.ylabel('Percentage of missing values')
plt.xlabel('Features')
plt.title('Missing Data in Percentages');
```

```python
for column in df.select_dtypes(include=['number']):
    df[column] = df[column].fillna(df[column].mean())

df
```
